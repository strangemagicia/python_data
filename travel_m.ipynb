```python
# 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('travel_rate.csv')

# 데이터를 확인
print(data.head(74))

# 데이터 준비
korea_travelers = data['korea-travel']
yen_rate = data['yen-rate-d']
usd_rate = data['won-rate-d']

# 히스토그램 생성
plt.figure(figsize=(15, 8))

# 1. 한국 여행자 수 히스토그램
plt.subplot(3, 1, 1)
plt.hist(korea_travelers, bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Number of Travelers')
plt.ylabel('Frequency')
plt.title('Distribution of Korea Travelers')

# 2. 원-엔 환율 히스토그램
plt.subplot(3, 1, 2)
plt.hist(yen_rate, bins=15, color='orange', alpha=0.7, edgecolor='black')
plt.xlabel('Exchange Rate (JPY to KRW)')
plt.ylabel('Frequency')
plt.title('Distribution of Yen Exchange Rates')

# 3. 원-달러 환율 히스토그램
plt.subplot(3, 1, 3)
plt.hist(usd_rate, bins=15, color='green', alpha=0.7, edgecolor='black')
plt.xlabel('Exchange Rate (USD to KRW)')
plt.ylabel('Frequency')
plt.title('Distribution of USD Exchange Rates')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
# "travel_rate.csv" 파일이 같은 디렉토리에 있어야 합니다.
data = pd.read_csv('travel_rate.csv')

# 데이터를 확인
print(data.head(74))

# 월별 데이터 설정
months = data['month']
korea_travelers = data['korea-travel']
yen_rate = data['yen-rate-d']
usd_rate = data['won-rate-d']

# 그래프 생성
plt.figure(figsize=(15, 8))

# 1. 해외여행자수 그래프 (한국 여행자 기준)
plt.subplot(2, 1, 1)
plt.plot(months, korea_travelers, color='blue', label='Korea Travelers')
plt.xlabel('Month')
plt.ylabel('Number of Travelers')
plt.title('Monthly Korea Travelers')
plt.xticks(rotation=45)
plt.legend()

# 2. 환율 그래프 (원-달러, 원-엔 환율 비교)
plt.subplot(2, 1, 2)
plt.plot(months, yen_rate, color='orange', label='Yen Rate (JPY to KRW)')
plt.plot(months, usd_rate, color='green', label='USD Rate (USD to KRW)')
plt.xlabel('Month')
plt.ylabel('Exchange Rate')
plt.title('Monthly Exchange Rates')
plt.xticks(rotation=45)
plt.legend()

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('travel_rate.csv')

# 데이터를 확인
print(data.head())

# 월별 데이터 설정
months = data['month']
korea_travelers = data['korea-travel']
yen_rate = data['yen-rate-d']
usd_rate = data['won-rate-d']

# 그래프 생성
plt.figure(figsize=(15, 15))

# 1. 월별 한국 여행자 수 막대 그래프
plt.subplot(3, 1, 1)
plt.bar(months, korea_travelers, color='blue', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Number of Travelers')
plt.title('Monthly Korea Travelers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 2. 월별 원-엔 환율 막대 그래프
plt.subplot(3, 1, 2)
plt.bar(months, yen_rate, color='orange', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Exchange Rate (JPY to KRW)')
plt.title('Monthly Yen Exchange Rates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 3. 월별 원-달러 환율 막대 그래프
plt.subplot(3, 1, 3)
plt.bar(months, usd_rate, color='green', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Exchange Rate (USD to KRW)')
plt.title('Monthly USD Exchange Rates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 전체 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
