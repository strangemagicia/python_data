```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import matplotlib.pyplot as plt

# CSV íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
# "travel_rate.csv" íŒŒì¼ì´ ê°™ì€ ë””ë ‰í† ë¦¬ì— ìžˆì–´ì•¼ í•©ë‹ˆë‹¤.
data = pd.read_csv('travel_rate.csv')

# ë°ì´í„°ë¥¼ í™•ì¸
print(data.head(74))

# ì›”ë³„ ë°ì´í„° ì„¤ì •
months = data['month']
korea_travelers = data['korea-travel']
yen_rate = data['yen-rate-d']
usd_rate = data['won-rate-d']

# ê·¸ëž˜í”„ ìƒì„±
plt.figure(figsize=(15, 8))

# 1. í•´ì™¸ì—¬í–‰ìžìˆ˜ ê·¸ëž˜í”„ (í•œêµ­ ì—¬í–‰ìž ê¸°ì¤€)
plt.subplot(2, 1, 1)
plt.plot(months, korea_travelers, color='blue', label='Korea Travelers')
plt.xlabel('Month')
plt.ylabel('Number of Travelers')
plt.title('Monthly Korea Travelers')
plt.xticks(rotation=45)
plt.legend()

# 2. í™˜ìœ¨ ê·¸ëž˜í”„ (ì›-ë‹¬ëŸ¬, ì›-ì—” í™˜ìœ¨ ë¹„êµ)
plt.subplot(2, 1, 2)
plt.plot(months, yen_rate, color='orange', label='Yen Rate (JPY to KRW)')
plt.plot(months, usd_rate, color='green', label='USD Rate (USD to KRW)')
plt.xlabel('Month')
plt.ylabel('Exchange Rate')
plt.title('Monthly Exchange Rates')
plt.xticks(rotation=45)
plt.legend()

# ê·¸ëž˜í”„ ë³´ì—¬ì£¼ê¸°
plt.tight_layout()
plt.show()
```

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import matplotlib.pyplot as plt

# CSV íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
data = pd.read_csv('travel_rate.csv')

# ë°ì´í„°ë¥¼ í™•ì¸
print(data.head())

# ì›”ë³„ ë°ì´í„° ì„¤ì •
months = data['month']
korea_travelers = data['korea-travel']
yen_rate = data['yen-rate-d']
usd_rate = data['won-rate-d']

# ê·¸ëž˜í”„ ìƒì„±
plt.figure(figsize=(15, 15))

# 1. ì›”ë³„ í•œêµ­ ì—¬í–‰ìž ìˆ˜ ë§‰ëŒ€ ê·¸ëž˜í”„
plt.subplot(3, 1, 1)
plt.bar(months, korea_travelers, color='blue', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Number of Travelers')
plt.title('Monthly Korea Travelers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 2. ì›”ë³„ ì›-ì—” í™˜ìœ¨ ë§‰ëŒ€ ê·¸ëž˜í”„
plt.subplot(3, 1, 2)
plt.bar(months, yen_rate, color='orange', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Exchange Rate (JPY to KRW)')
plt.title('Monthly Yen Exchange Rates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 3. ì›”ë³„ ì›-ë‹¬ëŸ¬ í™˜ìœ¨ ë§‰ëŒ€ ê·¸ëž˜í”„
plt.subplot(3, 1, 3)
plt.bar(months, usd_rate, color='green', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Exchange Rate (USD to KRW)')
plt.title('Monthly USD Exchange Rates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# ì „ì²´ ê·¸ëž˜í”„ ë ˆì´ì•„ì›ƒ ì¡°ì •
plt.tight_layout()

# ê·¸ëž˜í”„ ë³´ì—¬ì£¼ê¸°
plt.show()
```
